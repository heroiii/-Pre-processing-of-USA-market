```python
import pandas as pd
import numpy as np
import seaborn as sns
from matplotlib import pyplot as plt
```

```python
data = pd.read_csv('Market.csv')
df = pd.DataFrame(data)
df
```

```python
filt = df['Index'] == 'NYA'
filt
```

```python
df1 = df.loc[filt, :]
df1
```

```python
df1.dtypes
```

```python
df1.describe()
```

```python
float_cols =['Open','High','Low','Close','Adj Close','Volume']
df1[float_cols] = df1[float_cols].round().astype('Int64')
df1.dtypes
```

```python
df1
```

```python
df1['Date'] = pd.to_datetime(df1['Date'])
df1['Date']
```

```python
df1.sort_values(by='Date' , inplace = True)

df1
```

```python
df2 = df1.dropna(subset =['Date','Adj Close'] ,how = 'any')
df2
```
